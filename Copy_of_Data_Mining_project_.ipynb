## Data Visualization 

A histogram is typically used to visualize the distribution of a numerical variable. However, if you have a non-numerical column and you still want to visualize its distribution, you can convert the non-numerical categories into numerical representations and then create a histogram. One common way to achieve this is by using bar plots.

Data science offers a variety of fascinating plots and visualizations that can provide valuable insights from your data. Let's explore some examples:

        Violin Plot: Similar to a box plot, the violin plot displays the distribution of a numerical variable across different categories. It reveals the shape and density of the distribution.

        Swarm Plot: A swarm plot shows the distribution of data points for a categorical variable. It can be used alone or combined with a violin plot to visualize individual data points.

        Pair Plot: A pair plot, also known as a scatter matrix, creates a grid of scatter plots for multiple numerical variables. It helps uncover relationships between pairs of variables.


        Word Cloud: A word cloud represents text data visually, where the size of each word corresponds to its frequency in the dataset. It's commonly used in text analysis and sentiment analysis.


        Heat Map: A heat map displays data using a color-coded matrix, where each cell represents a value. It's often used to reveal correlations, patterns, or clustering in datasets.

## Data distribution plot on numerical and category columns 

```
 


import matplotlib.pyplot as plt

# Categorical variables
categorical_vars = ['SECTOR', 'METHOD', 'DATA SENSITIVITY']

# Create subplots for each categorical variable
fig, axes = plt.subplots(nrows=len(categorical_vars), figsize=(8, 6 * len(categorical_vars)))

# Define colors for the categories of DATA SENSITIVITY
data_sensitivity_colors = {
    'Just email address/Online information': 'gray',
    'SSN/Personal details': 'gray',
    'Credit card information': 'gray',
    'Health & other personal records': 'gray',
    'Full details': 'gray'
}

# Iterate over each categorical variable and create a bar plot
for i, var in enumerate(categorical_vars):
    # Count the occurrences of each category
    category_counts = df_copy[var].value_counts()

    # Sort the categories based on their counts
    sorted_categories = category_counts.index

    # Get the corresponding counts for each category
    counts = category_counts.values

    # Create the bar plot
    if var == 'DATA SENSITIVITY':
        # Use custom colors for DATA SENSITIVITY
        colors = [data_sensitivity_colors.get(category, 'gray') for category in sorted_categories]
        axes[i].bar(sorted_categories, counts, color=colors, label=var)
    else:
        # Use default colors for other variables
        axes[i].bar(sorted_categories, counts, label=var)

    # Add labels and title
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Count')
    axes[i].set_title(f'Distribution of {var}')

    # Rotate x-axis labels for 'SECTOR' variable
    if var == 'SECTOR':
        axes[i].set_xticklabels(sorted_categories, rotation=90)

    # Remove 'interesting story' subplot
    if var == 'interesting story':
        axes[i].remove()

    # Replace the x-axis labels for the 'DATA SENSITIVITY' variable
    if var == 'DATA SENSITIVITY':
        plt.setp(axes[i].get_xticklabels(), verticalalignment='bottom')

# Create a custom legend for DATA SENSITIVITY
data_sensitivity_legend = [
    plt.Rectangle((0, 0), 1, 1, color='gray') for _ in data_sensitivity_colors.values()
]
data_sensitivity_labels = ['1 - Just email address/Online information',
                           '2 - SSN/Personal details',
                           '3 - Credit card information',
                           '4 - Health & other personal records',
                           '5 - Full details']
plt.legend(data_sensitivity_legend, data_sensitivity_labels, title='Data Sensitivity')

# Adjust the spacing between subplots
plt.tight_layout(pad=3)

# Show the plots
plt.show()
```

## Area Plot 

### Interpretation: The area plot illustrates the cumulative records lost over time. The x-axis represents the years, and the y-axis represents the cumulative sum of records lost. This plot helps identify any trends or patterns in the cumulative records lost over the years.

```
# Importing the necessary libraries
import matplotlib.pyplot as plt

# Grouping the data by year and calculating the cumulative sum of records lost
data_by_year = df_copy.groupby('YEAR')['records lost'].sum().cumsum()

# Creating the area plot
plt.fill_between(data_by_year.index, data_by_year)
plt.xlabel('Year')
plt.ylabel('Cumulative Records Lost')
plt.title('Cumulative Records Lost over Time')

# Rotating x-axis labels
plt.xticks(rotation=90)

plt.show()

```
